```python
def printMatrix(A):
    for i in A:
        for j in i:
            print(j, end=" ")
        print()
```

```python
from PIL import Image, ImageOps
import numpy as np

img = Image.open("canvas.png")
img = ImageOps.grayscale(img)

A = np.asarray(img)
print(np.shape(A))
```

```python
img
```

```python
X = [[0, -1, 0], 
     [-1, 5, -1], 
     [0, -1, 0]]

def matrixConvulation(a, b):
    c = 0
    for i in range(-1, 2):
        for j in range(-1, 2):
            c += (X[1+i][1+j])*(A[a+i][b+j])
    return c
```

```python
output = np.zeros(np.shape(A), int)

for i in range(len(A)):
    for j in range(len(A[0])):
        if i == 0 or i == len(A)-1 or j == 0 or j == len(A[0])-1:
            output[i][j] = A[i][j]
        else:
            output[i][j] = matrixConvulation(i, j)
```

```python
result = Image.fromarray(output.astype(np.uint8), 'L')
```
